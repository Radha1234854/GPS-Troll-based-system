```python
import matplotlib.pyplot as plt
import random
from datetime import datetime

# Define vehicle types and their characteristics
class Vehicle:
    def __init__(self, vehicle_type, base_toll_rate, location):
        self.vehicle_type = vehicle_type
        self.base_toll_rate = base_toll_rate
        self.location = location
    
    def calculate_toll(self, distance, current_time):
        # Basic toll calculation (can be expanded)
        toll_charged = self.base_toll_rate * distance
        
        # Adjust toll based on time (e.g., night surcharge)
        if 22 >= current_time.hour >= 6:
            toll_charged *= 1.2  # Night surcharge
        
        return toll_charged

# Define locations and coordinates
locations = {
    "Mumbai": (19.0760, 72.8777),
    "Bangalore": (12.9716, 77.5946),
    "Delhi": (28.7041, 77.1025),
    "Chennai": (13.0827, 80.2707),
    "Kolkata": (22.5726, 88.3639),
    "Hyderabad": (17.3850, 78.4867),
    "Ahmedabad": (23.0225, 72.5714),
    "Pune": (18.5204, 73.8567),
    "Jaipur": (26.9124, 75.7873),
    "Lucknow": (26.8467, 80.9462)
}

# Sample vehicles
vehicles = [
    Vehicle("Car", 5, random.choice(list(locations.keys()))),
    Vehicle("Bus", 10, random.choice(list(locations.keys()))),
    Vehicle("Truck", 15, random.choice(list(locations.keys()))),
    Vehicle("Motorcycle", 3, random.choice(list(locations.keys()))),
    Vehicle("Van", 8, random.choice(list(locations.keys()))),
    Vehicle("Ambulance", 0, random.choice(list(locations.keys())))  # Example of toll-exempt vehicle
]

# Simulate toll calculation for each vehicle
def simulate_toll(vehicle, distance, current_time):
    toll = vehicle.calculate_toll(distance, current_time)
    print(f"Vehicle Type: {vehicle.vehicle_type}")
    print(f"Location: {vehicle.location}")
    print(f"Distance: {distance} km")
    print(f"Current Time: {current_time.strftime('%H:%M')}")
    print(f"Toll Charged: ₹ {toll:.2f}\n")
    return vehicle.location, toll

# Sample data for 10 vehicles
vehicle_numbers = ["MH01AB1234", "KA02CD5678", "DL03EF9101", "TN04GH1121", "UP05IJ3141",
                   "GJ06KL5161", "AP07MN7181", "TS08OP9201", "WB09QR1221", "HR10ST3241"]

fig, ax = plt.subplots(figsize=(10, 8))

# Plot locations
for loc, coord in locations.items():
    ax.scatter(coord[1], coord[0], s=100, label=loc, alpha=0.7)

# Simulate toll calculation and plot
for i in range(10):
    vehicle = random.choice(vehicles)
    distance = random.randint(10, 100)  # Random distance between 10 to 100 km
    current_time = datetime.now()  # Simulate current time
    location, toll = simulate_toll(vehicle, distance, current_time)
    ax.text(locations[location][1] + 0.1, locations[location][0], f"{vehicle.vehicle_type}\nToll: ₹ {toll:.2f}", fontsize=9)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('GPS Toll-based System Simulation')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.grid(True)
plt.show()
```
